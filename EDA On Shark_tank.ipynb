```python
Ashneer_present=len(st[st.AshneerGrover_present==1])
Anupam_present=len(st[st.AnupamMittal_present==1])
Aman_present=len(st[st.AmanGupta_present==1])
Namita_present=len(st[st.NamitaThapar_present==1])
Vinneta_present=len(st[st.VineetaSingh_present==1])
Peyush_present=len(st[st.PeyushBansal_present==1])
Ghazal_present=len(st[st.GhazalAlagh_present==1])
```

```python
fig=plt.figure(figsize=(8,4))
present=[Ashneer_present,Anupam_present,Aman_present,Namita_present,Vinneta_present,Peyush_present,Ghazal_present]
name=['Ashneer','Anupam','Aman','Namita','Vinneta','Peyush','Ghazal']
plt.bar(name,present,color='blue',width=0.4)
plt.xlabel("Shark Name")
plt.ylabel("Candidate Present")
plt.title("Candidate Present vs SharkName")
plt.show()
```

```python
plt.pie(present,labels=name,radius=1,autopct='%0.1f%%')
plt.suptitle('% of Candidates present ',x=0.55,y=1.15)
```
